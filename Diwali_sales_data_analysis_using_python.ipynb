## Exploratory Data Analysis

## Male v/s Female purchased status

```python
plt.bar(df.Gender.unique(),df.Gender.value_counts(),color=['g','y'])
plt.title('Male v/s Female purchased status')
plt.xlabel('Gender')
plt.ylabel('Number of purchases')


for g,f in zip(df.Gender.unique(),df.Gender.value_counts(),):
    plt.text(g,f,f"{f} pics bought.")

```

## Male v/s Female spent money

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_status = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_status, x='Gender',y="Amount")
# Create a bar plot using Seaborn
sales_status = sns.barplot(data=sales_status, x='Gender', y='Amount')
sales_status.bar_label(sales_status.containers[0])

plt.show()
```

*From the above graphs we can identify that most of the buyeres are females and even the purchagin power of females are greater than men*

## Age group wise sales distribution

```python
df.columns
```

```python
agegroups=sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in agegroups.containers:
    agegroups.bar_label(bars)

plt.title("Number of products purchaged by Male/Female")
plt.show()
```

*By this chart we can see compare that most of the buyers age of group 26-35 years female*

## Age group wise revenue generated

```python
age_revenue = df.groupby(['Age Group'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
palette=sns.color_palette('husl',len(age_revenue))
sns.barplot(data=age_revenue,x='Age Group',y='Orders',palette=palette) 

plt.title("Age Group wise revenue generated")

plt.show()
```

*From the above Chart we can identify that most of the buyers are of age group between 26-35 years female*

## State wise orders distribution

```python
Sales_states = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)[:10]

palette=sns.color_palette("husl",10)
sns.barplot(data=Sales_states, x='State', y= 'Orders',palette=palette)

plt.title("States wise orders")
plt.show()
```

```python
Sales_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)[:10]
Palette=sns.color_palette("husl",10)
sns.barplot(data=Sales_states, x='State', y= 'Amount',palette=Palette)

plt.ylabel('Amount in thousands')

plt.title("States wise revenue generated")
plt.show()
```

*From the above 2 charts we can see the most of the Orders & Total sales/amount are from Utter Pradesh, Maharastra and Kamataka respectively..*

# Occupation wise __

```python
sns.set(rc={'figure.figsize':(20,5)})
palette = sns.color_palette("husl",15)
ax=sns.countplot(data=df, x='Occupation',palette=palette)
plt.title("Number of Quantity purchased")

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_states=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

# Generate a palette with 15 colors
palette = sns.color_palette("husl", 15)

# Plotting with different colors for each bin
sns.barplot(data=sales_states, x='Occupation', y='Amount', palette=palette)

plt.title("Occupation wise revenue generated")
plt.show()
```

*From the above chart we can see tht most of the buyers are working in IT, Healthcare and Aviation sector*

# Product category wise insides__

```python
pal=sns.color_palette('husl')
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(data=df, x='Product_Category',palette=pal)
for bars in ax.containers:
    ax.bar_label(bars)
```

*From this chart we can find the most of the people spend their on Food, Electronics & Gadgets and Clothing & Apparel*

```python
Prod_wise_revenue=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
pcolor=sns.color_palette('husl')
sns.barplot(data=Prod_wise_revenue, x='Product_Category', y='Amount',palette=pcolor)
plt.show()
```

 *From the above graph we can see that most of the sold products are food, Clothing & Apparel and Electronic*

```python
# Top most sold products__
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
plt.title("Top 10 sold products")
plt.show()
```
